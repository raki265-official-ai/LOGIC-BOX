# 💠WELCOME TO THE LOGIC BOX

```python
def pattern_generator():
    rows = int(input("Enter number of rows: "))

    if rows <= 0:
        print("Invalid rows!")
        return

    print("\nPattern:")
    for i in range(1, rows + 1):
        for j in range(i):
            print("*", end="")
        print()


def number_analyzer():
    start = int(input("Enter start number: "))
    end = int(input("Enter end number: "))

    if end < start:
        print("Invalid range!")
        return

    total = 0

    for num in range(start, end + 1):
        if num % 2 == 0:
            print(num, "is Even")
        else:
            print(num, "is Odd")

        total += num

    print("Sum =", total)


while True:
    print("\n--- Logic Box Menu ---")
    print("1. Generate Pattern")
    print("2. Analyze Numbers")
    print("3. Exit")

    choice = input("Enter choice: ")

    if choice == '1':
        pattern_generator()

    elif choice == '2':
        number_analyzer()

    elif choice == '3':
        print("Goodbye!")
        break

    else:
        print("Invalid Choice")
```
